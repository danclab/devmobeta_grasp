```python
import json
```

```python
files = ["bursts_three_BV.json", "bursts_three_EGI.json"]

merged_data_three = []

for file in files: 
    with open(file, 'r') as f:
        data = json.load(f)  # Charge le contenu du fichier
        if isinstance(data, list): 
            merged_data_three.extend(data)
        else:
            print(f"erreur : le fichier {file} ne contient pas une lsite.")

# Sauvegarde dans un nouveau fichier
output_file = "merged_bursts_three.json"
with open(output_file, 'w') as f_out:
    json.dump(merged_data_three, f_out, indent=4)

print(f"Les fichiers ont été fusionnés dans {output_file}.")
```
