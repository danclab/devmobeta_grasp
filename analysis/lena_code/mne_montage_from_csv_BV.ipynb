```python
import mne
import pandas as pd
import numpy as np

df = pd.read_csv('/home/common/bonaiuto/devmobeta/derivatives/montage_BV_64.csv', delimiter=',')

# Create a dictionary of positions
# Coordinates must be in (x, y, z) format
# We first need to convert spherical coordinates (phi, theta) to Cartesian coordinates

# Convert degrees to radians if your angles are in degrees
phi_rad = np.radians(df['Phi'].values)
theta_rad = np.radians(df['Theta'].values)

# Conversion of spherical coordinates to Cartesian coordinates
# Assuming that r = 1 (unit radius)
x = np.sin(theta_rad) * np.cos(phi_rad)
y = np.sin(theta_rad) * np.sin(phi_rad)
z = np.cos(theta_rad)

# Create a dictionary {channel_name: [x, y, z]}
ch_pos = {}
for i, name in enumerate(df['Name']):
    ch_pos[name] = np.array([x[i], y[i], z[i]])

# Create the assembly with the position dictionary
montage = mne.channels.make_dig_montage(ch_pos=ch_pos)

# Save the edit in .sfp format
montage.save('/home/common/bonaiuto/devmobeta/derivatives/montage_BV_64.sfp')
```
